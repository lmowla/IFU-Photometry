```python
def twoD_Gaussian((x, y), amplitude, x1, y1, x2, y2, sigma_x, sigma_y, theta, offset):
    xo = float(x1)
    yo = float(y1)    
    a = (np.cos(theta)**2)/(2*sigma_x**2) + (np.sin(theta)**2)/(2*sigma_y**2)
    b = -(np.sin(2*theta))/(4*sigma_x**2) + (np.sin(2*theta))/(4*sigma_y**2)
    c = (np.sin(theta)**2)/(2*sigma_x**2) + (np.cos(theta)**2)/(2*sigma_y**2)
    g = offset + amplitude*np.exp( - (a*((x-x1)**2) + 2*b*(x-x1)*(y-y1) 
                            + c*((y-y1)**2))) + amplitude*np.exp( - (a*((x-x2)**2) + 2*b*(x-x2)*(y-y2) 
                            + c*((y-y2)**2)))
    return g.ravel()
```

```python
# Create x and y indices
p = np.linspace(0, 400, 201)
q = np.linspace(0, 200, 201)
r = np.meshgrid(p, q)

#create data
disk = twoD_Gaussian(r, 1, 150, 100,300,100, 20, 20, 0, 10)
peak = twoD_Gaussian(r, 5, 150, 100,300,100, 5, 5, 0, 10)
comb = disk + peak

# Contours
#cnt = ax.contour(comb.reshape(201, 201), colors='k', vmin=abs(comb).min(), vmax=abs(comb).max())

# plot twoD_Gaussian data generated above
plt.figure()
plt.imshow(comb.reshape(201, 201))
plt.colorbar()
```

```python
import numpy as np
from scipy.stats import multivariate_normal

x, y = np.mgrid[1:100:100j,1:200:200j]
# Need an (N, 2) array of (x, y) pairs.
xy = np.column_stack([x.flat, y.flat])

mu11 = np.array([30, 120])
sigma11 = np.array([2, 4])
covariance11 = np.diag(sigma11**2)

mu12 = np.array([30, 120])
sigma12 = np.array([5, 10])
covariance12 = np.diag(sigma12**2)

mu21 = np.array([30, 80])
sigma21 = np.array([2, 4])
covariance21 = np.diag(sigma21**2)

mu22 = np.array([30, 80])
sigma22 = np.array([5, 10])
covariance22 = np.diag(sigma22**2)

z11 = 1.5*multivariate_normal.pdf(xy, mean=mu11, cov=covariance11)
z12 = 10*multivariate_normal.pdf(xy, mean=mu12, cov=covariance12)
z1 = z11 + z12

z21 = multivariate_normal.pdf(xy, mean=mu21, cov=covariance21)
z22 = 10*multivariate_normal.pdf(xy, mean=mu22, cov=covariance22)
z2 = z21 + z22

z = z1 + z2

# Reshape back to a (30, 30) grid.
#z = z.reshape(x.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Intensity')

plt.title('Gaussian Model for KOI13')

plt.savefig('KOI13_Model.png',dpi=300)
```

```python
# Voigt profile

from scipy import special
N = 10
a = linspace
def voigt(x,amp,pos,fwhm,shape):
     """\
voigt profile

V(x,sig,gam) = Re(w(z))/(sig*sqrt(2*pi))
z = (x+i*gam)/(sig*sqrt(2))
     """

     tmp = 1/special.wofz(N.zeros((len(x))) \
           +1j*N.sqrt(N.log(2.0))*shape).real
     tmp = tmp*amp* \
           special.wofz(2*N.sqrt(N.log(2.0))*(x-pos)/fwhm+1j* \
           N.sqrt(N.log(2.0))*shape).real
     return tmp
```
