```python
xy = np.column_stack([x.flat, y.flat])

mu11 = np.array([DEC_max[0], RA_max[0]])
sigma11 = np.array([10, 20])
covariance11 = np.diag(sigma11**2)

mu12 = np.array([DEC_max[0], RA_max[0]])
sigma12 = np.array([40, 150])
covariance12 = np.diag(sigma12**2)

mu21 = np.array([DEC_max[0], RA_max[0]+500])
sigma21 = np.array([10, 20])
covariance21 = np.diag(sigma21**2)

mu22 = np.array([DEC_max[0], RA_max[0]+500])
sigma22 = np.array([40, 150])
covariance22 = np.diag(sigma22**2)

z11 = 1.5*multivariate_normal.pdf(xy, mean=mu11, cov=covariance11)
z12 = 20*multivariate_normal.pdf(xy, mean=mu12, cov=covariance12)
z1 = z11 + z12

z21 = 2.5*multivariate_normal.pdf(xy, mean=mu21, cov=covariance21)
z22 = 20*multivariate_normal.pdf(xy, mean=mu22, cov=covariance22)
z2 = z21 + z22

z = z1 + z2

# Reshape back to a (30, 30) grid.
z = z.reshape(x.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

p = ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)

ax.set_xlabel('RA J2000.0')
ax.set_ylabel('DEC J2000.0')
ax.set_zlabel('Intensity')

fig.colorbar(p,ax=ax)

plt.title('Gaussian Model for KOI13')

plt.savefig('KOI13_Model.png',dpi=300)
```

```python
def two_voigt (v1_mu_ra, v1_mu_dec, v2_mu_ra, v2_mu_dec,sig1,sig2,gam1,gam2, A1=100, A2=150):
    #voigt=VoightModel()
    x,y = np.mgrid[DEC_init:DEC_fin:increment_dec, RA_init:RA_fin:increment_ra]
    v1_ra=VoigtModel.f(voigt,x,A1,sig1,gam1,v1_mu_ra)
    v1_dec=VoigtModel.f(voigt,y,A1,sig1,gam1,v1_mu_dec).transpose()
    v2_ra=VoigtModel.f(voigt,x,A2,sig2,gam2,v2_mu_ra)
    v2_dec=VoigtModel.f(voigt,y,A2,sig2,gam2,v2_mu_dec).transpose()
    v = v1_ra*v1_dec + v2_ra*v2_dec
    #plot_voigt = plt.imshow(v)
    return v
```
